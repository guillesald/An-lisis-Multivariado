## Dataset Escuelas Primarias ##


```r
#filtrar datos y generar nuevo dataset para las escuelas primarias
dim(NOGOYA)

NOGOYAP <-filter(NOGOYA,  NivelEnsenanza=="PRIMARIO" & EsMultianio== "N" &  ModEnsenanza=="Comun")

dim(NOGOYAP)
NOGOYAP
```

```r
table(NOGOYAP$Turno)
```

```r
# Contar el número total de códigos únicos (escuelas)
num_total_escuelas_p <- NOGOYAP %>%
  summarise(num_total_escuelas_p = n_distinct(CUE))

# Mostrar el resultado
print(num_total_escuelas_p)
```

### Procesamiento de la variable Curso ###

```r
table(NOGOYAP$Curso)
```

```r
NOGOYAP2 <-filter(NOGOYAP, !grepl("y", Curso) & !grepl("Y", Curso) & !grepl("a", Curso) & !grepl("°", Curso) & !grepl(",", Curso) & !grepl("º", Curso))
table(NOGOYAP2$Curso)
```

```r
NOGOYAP3 <-filter(NOGOYAP2, !grepl("4-5-6 - A Prim", Curso) & !grepl("3- 4-5-6 - A Prim", Curso) & !grepl("1 - 6 - A Prim", Curso))
table(NOGOYAP3$Curso)
```

```r
NOGOYAP3$Curso <- substr(NOGOYAP3$Curso, 1, 1)
table(NOGOYAP3$Curso)
```

### Total de alumnos ###

```r
# Contar el número total de documentos únicos (alumnos)
num_total_alumnos_p <- NOGOYAP3 %>%
  summarise(num_total_alumnos_p = n_distinct(Documento))

# Mostrar el resultado
print(num_total_alumnos_p)
```

### Separacion por Trimestre ###

```r
PrimerTPrimario<-filter(NOGOYAP3,PeriodoEvaluatorio=="Primer Trimestre")
head(PrimerTPrimario)
names(PrimerTPrimario)
PrimerTPrimario<-rename(PrimerTPrimario,NotasPrimerTPrimario=Nota)
dim(PrimerTPrimario)
names(PrimerTPrimario)
```

```r
SegundoTPrimario<-filter(NOGOYAP3,PeriodoEvaluatorio=="Segundo Trimestre")
head(SegundoTPrimario)
names(SegundoTPrimario)
SegundoTPrimario<-rename(SegundoTPrimario,NotasSegundoTPrimario=Nota)
dim(SegundoTPrimario)
names(SegundoTPrimario)
```

```r
TercerTPrimario<-filter(NOGOYAP3,PeriodoEvaluatorio=="Tercer Trimestre")
head(TercerTPrimario)
names(TercerTPrimario)
TercerTPrimario<-rename(TercerTPrimario,NotasTercerTPrimario=Nota)
dim(TercerTPrimario)
names(TercerTPrimario)
```
